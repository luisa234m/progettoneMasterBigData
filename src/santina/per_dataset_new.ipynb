```python
def get_population_for_olympic_year(country_name, olympic_year, pop_data, years_back=4):
    target_years = list(range(olympic_year - years_back, olympic_year))
    subset = pop_data[
        (pop_data['Country Name'] == country_name) &
        (pop_data['year'].isin(target_years))
    ]['population_wb']
    if subset.dropna().empty:
        return np.nan
    return subset.mean(skipna=True)


nan_mask = df_socioec['Popolazione_Totale'].isna()

df_socioec.loc[nan_mask, 'Popolazione_Totale'] = df_socioec[nan_mask].apply(
    lambda row: get_population_for_olympic_year(
        country_name=row['Nazione'],
        olympic_year=int(row['Anno']),
        pop_data=pop_long
    ),
    axis=1
)
```
