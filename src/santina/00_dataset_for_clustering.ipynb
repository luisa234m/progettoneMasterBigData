```python
#Funzione per calcolare il tipo e il numero di medaglie
def medals(group):
    return pd.Series({
        'gold':   (group['medal'] == 'Gold').sum(),
        'silver': (group['medal'] == 'Silver').sum(),
        'bronze': (group['medal'] == 'Bronze').sum(),
    })
```

## Dataset "Macro_sport"

```python
#Creo il DataFrame "Macro_sport" che contiene, per ogni olimpiade:
# - country_noc: codice noc dei paesi,
# - country: i nomi estesi dei paesi partecipanti
# - year: l'anno dell'olimpiade
# - sport: la categoria sportiva
# - num_athletes: il numero di atleti partecipanti in ogni categoria sportiva per paese e anno
# - old, silver, bronze: il numero di medaglie d'oro, di bronzo, d'argento
# - total_medals: il numero di medaglie totali
# - gold_per_athletes: medaglie d'oro per numero di atleti partecipanti
# - tot_med_per_athletes: medaglie totali per numero di atleti partecipanti

df_macro_sport = df_clean.groupby(['country_noc', 'year', 'sport', 'num_athletes']).apply(
    medals
).reset_index()

df_macro_sport['total_medals'] = (df_macro_sport['gold'] + df_macro_sport['silver'] + df_macro_sport['bronze'])
df_macro_sport['gold_per_athletes'] = df_macro_sport['gold'] / df_macro_sport['num_athletes']
df_macro_sport['tot_med_per_athletes'] = df_macro_sport['total_medals'] / df_macro_sport['num_athletes']
df_macro_sport
```

```python
#Aggiungo al DataFrame il nome esteso del paese e salvo come dataset da usare in analisi successive
df_countries = df_countries.rename(columns={'noc': 'country_noc'})
df_countries['country'] = df_countries['country'].replace(country_mapping)
df_macro_sport = df_macro_sport.merge(df_countries, on='country_noc', how='left')
cols = list(df_macro_sport.columns)
cols.insert(cols.index('country_noc') + 1, cols.pop(cols.index('country')))
df_macro_sport = df_macro_sport[cols]
df_macro_sport = df_macro_sport.sort_values('country', key=lambda x: x.str.len(), ascending=False)
df_macro_sport = df_macro_sport.drop_duplicates(subset=['country_noc', 'year', 'sport'], keep='first')
df_macro_sport.to_csv("Macro_sports.csv", index=False)
df_macro_sport
```

## Dataset sugli sport nel tempo

```python
#Creo un dataset con le segeunti features:
# - sport: la categoria sportiva
# - year: anno dell'edizione olimpica
# - n_nations: numero di Paesi che hanno in quell'anno hanno partecipato ad eventi appartenenti a quella categoria sportiva
# - n_events: numero di eventi appartenenti a quella categoria sportiva
# - athletes_per_event: numero di atleti medio per evento nella categoria sportiva --> quanto è mediamente grande ogni evento sportivo nella categoria
# - pct_team_events: percentuali di eventi di squadra nella categoria sportiva (0 = tutti eventi individuali; 1 = tutti eventi a squadre)
# - hhi_gold: Herfindahl-Hirschman Index per medaglie d'oro --> Indice statistico che per la categoria sportiva misura la concentrazione delle medaglie d'oro tra i paesi, quindi la distribuzione delle medaglie d'oro tra i vari Paesi negli eventi sportivi appartenenti a quella categoria (se vicino a 1 un solo Paese vince tutti gli ori, se vicino a 0 gli ori sono distributi tra tanti paesi).
# - hhi_gold_athletes: Herfindahl-Hirschman Index per medaglie d'oro pesato per numero di atleti --> Paesi più efficienti in base al numero di atleti pesano di più
# - n_nations_with_gold: numero di nazioni che hanno vinto almeno una medaglia d'oro negli eventi di quella categoria sportiva
# - mean_gold_per_athletes: media degli ori per atleta tra i Paesi vincitori
# - hhi_total: Herfindahl-Hirschman Index per medaglie totali
# - hhi_total_athletes: Herfindahl-Hirschman Index per medaglie totali pesato per numero di atleti
# - n_nations_with_medal: numero di nazioni che hanno vinto almeno una medaglia negli eventi di quella categoria sportiva
# - mean_total_per_athletes: media delle medalie totali per atleta tra i Paesi vincitori di medaglia
# salvo il dataset per le analisi successive

sport_year = []

for (sport, year), grp in df_clean.groupby(['sport', 'year']):

    n_nations = grp['country_noc'].nunique()
    n_events = grp['event'].nunique()
    athletes_per_event = grp.groupby('event')['athlete_id'].nunique().mean()
    event_team_flag = grp.groupby('event')['isTeamSport'].first()
    pct_team_events = event_team_flag.mean()

    medals_grp = grp[grp['medal'].notna()]
    athletes_by_country = grp.groupby('country_noc')['num_athletes'].first()

    if len(medals_grp) > 0:

        gold_by_country = medals_grp[medals_grp['medal'] == 'Gold'].groupby('country_noc').size()
        total_gold = gold_by_country.sum()
        n_nations_with_gold = gold_by_country.shape[0]
        hhi_gold = ((gold_by_country / total_gold) ** 2).sum() if total_gold > 0 else np.nan
        gold_per_athl = (gold_by_country / athletes_by_country).dropna()
        hhi_gold_athl = ((gold_per_athl / gold_per_athl.sum()) **2).sum() if gold_per_athl.sum() > 0 else np.nan
        mean_gold_per_athl = gold_per_athl.mean()

        total_by_country = medals_grp.groupby('country_noc').size()
        total_medals = total_by_country.sum()
        n_nations_with_medal = total_by_country.shape[0]
        hhi_total = ((total_by_country / total_medals) ** 2).sum() if total_medals > 0 else np.nan
        total_per_athl = (total_by_country / athletes_by_country).dropna()
        hhi_total_athl = ((total_per_athl / total_per_athl.sum()) ** 2).sum() if total_per_athl.sum() > 0 else np.nan
        mean_total_per_athl = total_per_athl.mean()

    else:
        hhi_gold = np.nan
        hhi_gold_athl = np.nan
        n_nations_with_gold = 0
        mean_gold_per_athl = np.nan

        hhi_total = np.nan
        hhi_total_athl = np.nan
        n_nations_with_medal = 0
        mean_total_per_athl = np.nan

    sport_year.append({
        'sport': sport,
        'year': year,
        'n_nations': n_nations,
        'n_events': n_events,
        'athletes_per_event': athletes_per_event,
        'pct_team_events': pct_team_events,
        'hhi_gold': hhi_gold,
        'hhi_gold_athletes': hhi_gold_athl,
        'n_nations_with_gold': n_nations_with_gold,
        'mean_gold_per_athletes': mean_gold_per_athl,
        'hhi_total': hhi_total,
        'hhi_total_athletes': hhi_total_athl,
        'n_nations_with_medal': n_nations_with_medal,
        'mean_total_per_athletes': mean_total_per_athl,
    })

df_sport_year = pd.DataFrame(sport_year)
df_sport_year.to_csv("Sport_features_by_year.csv", index=False)
df_sport_year
```

## Dataset sul profilo di ogni sport (media sui vari anni)

```python
#Creo un dataset sul profilo generico di ogni categoria sportiva (media storica su tutte le edizioni), le features sono:
# - sport: categoria sportiva
# - n_nations_mean: numero medio storico di Paesi partecipanti nei vari anni negli eventi sportivi appartenenti a quella categoria sportiva
# - n_nations_max: numero massimo storico di Paesi partecipanti
# - athletes_per_event: media storica di atleti per evento sportivo appartenente a quella categoria sportiva
# - pct_team_events: numero medio di eventi sportivi a squadre in quella categoria sportiva
# - hhi_gold_mean: concentrazione media storica delle medaglie d'oro (Herfindahl-Hirschman Index)
# - hhi_gold_athletes_mean: concentrazione media storica degli ori pesata per numero di atleti
# - mean_gold_per_athletes: media storica del rapporto ori/atleti tra i paesi vincitori di oro
# - n_nations_with_gold: numero medio storico di paesi che hanno vinto almeno una medaglia d'oro in quella categoria sportiva
# - hhi_total_mean: concentrazione media storica delle medaglie totali tra i paesi
# - hhi_total_athletes_mean: concentrazione media storica delle medaglie totali pesata per numero di atleti
# - n_nations_with_medal: media storica del numero di paesi che hanno vinto almeno una medaglia
# - mean_total_per_athletes: media storica del rapporto medaglie totali/atleti tra i paesi medagliati
# - n_editions: numero di edizioni olimpiche in cui era presente quella categoria sportiva
# salvo il dataset per le analisi successive

df_sport_profile = df_sport_year.groupby('sport').agg(
    n_nations_mean = ('n_nations', 'mean'),
    n_nations_max = ('n_nations', 'max'),
    athletes_per_event = ('athletes_per_event', 'mean'),
    pct_team_events = ('pct_team_events', 'mean'),
    hhi_gold_mean = ('hhi_gold', 'mean'),
    hhi_gold_athletes_mean = ('hhi_gold_athletes', 'mean'),
    mean_gold_per_athletes = ('mean_gold_per_athletes', 'mean'),
    n_nations_with_gold = ('n_nations_with_gold', 'mean'),
    hhi_total_mean = ('hhi_total', 'mean'),
    hhi_total_athletes_mean = ('hhi_total_athletes', 'mean'),
    n_nations_with_medal = ('n_nations_with_medal', 'mean'),
    mean_total_per_athletes = ('mean_total_per_athletes', 'mean'),
    n_editions = ('year', 'nunique'),
).reset_index()

df_sport_profile.to_csv("Sport_profile_static.csv", index=False)
df_sport_profile
```

## Dataset sul successo olimpico

```python
# Creo un dataset che misura il successo olimpico storico di ogni paese in ogni categoria sportiva:
# - country_noc: codice noc del paese
# - sport: categoria sportiva
# - success_score_mean: media storica del rapporto ori/atleti per edizione olimpica
# - total_golds: numero totale di ori vinti storicamente in quella categoria sportiva
# - total_athletes: numero totale di atleti inviati storicamente in quella categoria sportiva
# - gold_per_athletes: ori totali / atleti totali storici (efficienza aggregata del paese in quello sport)
# salvo il dataset per le analisi successive

athletes_by_csy = df_clean.groupby(['country_noc', 'sport', 'year'])['athlete_id'].nunique().reset_index()
athletes_by_csy.columns = ['country_noc', 'sport', 'year', 'athletes_sent']

gold_by_csy = (
    df_clean[df_clean['medal'] == 'Gold']
    .groupby(['country_noc', 'sport', 'year'])
    .size()
    .reset_index(name='golds_won')
)

success_df = athletes_by_csy.merge(gold_by_csy, on=['country_noc', 'sport', 'year'], how='left')
success_df['golds_won'] = success_df['golds_won'].fillna(0)
success_df['success_score'] = success_df['golds_won'] / success_df['athletes_sent']

df_success_country_sport = success_df.groupby(['country_noc', 'sport']).agg(
    success_score_mean = ('success_score', 'mean'),
    total_golds = ('golds_won', 'sum'),
    total_athletes = ('athletes_sent', 'sum'),
).reset_index()

#ori totali / atleti totali storici: efficienza aggregata del paese in quello sport
df_success_country_sport['gold_per_athletes'] = (df_success_country_sport['total_golds'] / df_success_country_sport['total_athletes'])

df_success_country_sport.to_csv("Success_score_country_sport.csv", index=False)
df_success_country_sport
```

## Dataset sui singoli eventi

```python
#Creo un dataset con le seguenti features per ogni combinazione evento + anno:
# - sport: categoria sportiva a cui appartiene l'evento
# - event: nome dell'evento sportivo appartenente a quella categoria
# - year: anno dell'edizione olimpica
# - n_nations: numero di paesi partecipanti all'evento in quell'anno
# - n_athletes: numero totale di atleti partecipanti all'evento in quell'anno
# - is_team_event: indica se l'evento è a squadre o individuale
# - gold_per_athlete: media tra i paesi del rapporto ori/atleti --> efficienza normalizzata per dimensione del paese
# - n_nations_with_gold: numero di paesi che hanno vinto l'oro (in un singolo evento per anno è sempre 1)
# - total_per_athlete: media tra i paesi del rapporto medaglie totali/atleti
# - n_nations_with_medal: numero di paesi che hanno vinto almeno una medaglia
# salvo il dataset per le analisi successive

# In questo dataset non creo la feature sull'hhi perché in un singolo evento per anno c'è sempre un solo oro (l'hhi sarebbe sempre 1 e non aggiungerebbe informazione)

event_year_rows = []

for (sport, event, year), grp in df_clean.groupby(['sport', 'event', 'year']):

    n_nations  = grp['country_noc'].nunique()
    n_athletes = grp['athlete_id'].nunique()
    is_team    = grp['isTeamSport'].iloc[0]

    medals_grp = grp[grp['medal'].notna()]
    athletes_by_country = grp.groupby('country_noc')['num_athletes'].first()

    if len(medals_grp) > 0:

        gold_by_country = medals_grp[medals_grp['medal'] == 'Gold'].groupby('country_noc').size()
        n_nations_with_gold = gold_by_country.shape[0]
        gold_per_athlete = (gold_by_country / athletes_by_country).dropna().mean()

        total_by_country = medals_grp.groupby('country_noc').size()
        n_nations_with_medal = total_by_country.shape[0]
        total_per_athlete = (total_by_country / athletes_by_country).dropna().mean()

    else:
        n_nations_with_gold  = 0
        gold_per_athlete     = np.nan
        n_nations_with_medal = 0
        total_per_athlete    = np.nan

    event_year_rows.append({
        'sport':               sport,
        'event':               event,
        'year':                year,
        'n_nations':           n_nations,
        'n_athletes':          n_athletes,
        'is_team_event':       is_team,
        'gold_per_athlete':     gold_per_athlete,
        'n_nations_with_gold':  n_nations_with_gold,
        'total_per_athlete':    total_per_athlete,
        'n_nations_with_medal': n_nations_with_medal
    })

df_event_year = pd.DataFrame(event_year_rows)
df_event_year.to_csv("Event_features_by_year.csv", index=False)
df_event_year
```

## Dataset sul profilo di ogni evento sportivo (media sui vari anni)

```python
# Funzione per calcolare l'hhi su un gruppo di paesi, per calcolare se in quell'evento ha sempre vinto l'oro un unico paese o se nel corso del tempo hanno vinto paesi diversi
def hhi_from_group(g):
    total = g['golds_won'].sum()
    return ((g['golds_won'] / total) ** 2).sum() if total > 0 else np.nan
```

```python
#Creo un dataset sul profilo generico di ogni evento sportivo (media storica su tutte le edizioni):
# - sport: categoria sportiva a cui appartiene l'evento
# - event: nome dell'evento sportivo appartenente a quella categoria
# - is_team_event: indica se l'evento è a squadre o individuale
# - n_nations_mean: media storica del numero di paesi partecipanti
# - n_nations_max: massimo storico del numero di paesi partecipanti
# - n_athletes_mean: media storica del numero di atleti partecipanti
# - gold_per_athlete: media storica del rapporto ori/atleti tra i paesi vincitori
# - n_nations_with_gold: media storica del numero di paesi che hanno vinto l'oro
# - total_per_athlete: media storica del rapporto medaglie totali/atleti tra i paesi medagliati
# - n_nations_with_medal: media storica del numero di paesi che hanno vinto almeno una medaglia
# - n_editions: numero di edizioni olimpiche in cui è stato presente quell'evento
# - hhi_gold: concentrazione storica degli ori (misura se lo stesso paese vince sempre o se l'oro cambia mani nel tempo)
# - hhi_total: concentrazione storica delle medaglie totali tra i paesi

df_event_profile = df_event_year.groupby(['sport', 'event', 'is_team_event']).agg(
    n_nations_mean      = ('n_nations', 'mean'),
    n_nations_max       = ('n_nations', 'max'),
    n_athletes_mean     = ('n_athletes', 'mean'),
    gold_per_athlete        = ('gold_per_athlete', 'mean'),
    n_nations_with_gold     = ('n_nations_with_gold', 'mean'),
    total_per_athlete       = ('total_per_athlete', 'mean'),
    n_nations_with_medal    = ('n_nations_with_medal', 'mean'),
    n_editions          = ('year', 'nunique'),
).reset_index()

gold_winner_by_event = (
    df_clean[df_clean['medal'] == 'Gold']
    .groupby(['sport', 'event', 'country_noc'])
    .size()
    .reset_index(name='golds_won')
)

hhi_event = (
    gold_winner_by_event
    .groupby(['sport', 'event'])
    .apply(hhi_from_group)
    .reset_index(name='hhi_gold')
)

total_winner_by_event = (
    df_clean[df_clean['medal'].notna()]
    .groupby(['sport', 'event', 'country_noc'])
    .size()
    .reset_index(name='medals_won')
)

hhi_event_total = (
    total_winner_by_event
    .groupby(['sport', 'event'])
    .apply(lambda g: ((g['medals_won'] / g['medals_won'].sum()) ** 2).sum())
    .reset_index(name='hhi_total')
)

df_event_profile = df_event_profile.merge(hhi_event, on=['sport', 'event'], how='left')
df_event_profile = df_event_profile.merge(hhi_event_total, on=['sport', 'event'], how='left')

df_event_profile.to_csv("Event_profile_static.csv", index=False)

df_event_profile
```
